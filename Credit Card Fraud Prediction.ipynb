```python
#under sampling in n=473 fraud transction.
legit_sample = legit.sample(n=473)
#cooncating the new df.
new_df = pd.concat([legit_sample, fraud], axis=0)
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
#split the new_df.
x=new_df.drop(columns='Class',axis=1)
y=new_df['Class']
print(x)
print(y)
```

```python
x_train = new_df.drop(columns = 'Class', axis=1)
y_train = new_df['Class']
print(y_train.shape)
```

```python
#train_test_split the dataframe model.
x_train,y_train,x_test,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
print(x.shape,x_train.shape,x_test.shape)
```

```python
#model train by LogisticRegession.
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#accuracy_score by tarning model evaluation.
x_train_prediction = model.predict(x_train)
train_accuracy = accuracy_score(x_train_prediction,y_train)
print("Accuracy Score by Training data ::", train_accuracy)
```
